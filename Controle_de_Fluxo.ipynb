```python
cont = 10
while cont >= 0:
  print(cont,end=' ')
  cont = cont - 1
```

```python
while True:
  resposta = input()
  if resposta == '1':
    break
  else:
    print('Resposta invalida.')
```

```python
for numero in range(10):
  print(f'{numero}',end = ' ')
```

```python
for i in range(0,10,2):
  print(i)
```

```python
try:
  numerador = float(5)
  denominador = float(0)
  resultado = numerador / denominador
  print(f'o resultado da divisao e {resultado:.4f}')
except ZeroDivisionError:
  print('Erro: Nao e possivel dividir por Zero')
```

```python
for numero in range(1,11):
  print(f'Numero {numero}')
  for i in range(1,11):
    print(f'{numero} X {i} = {numero * 1}')
  print('-'*20)
```
