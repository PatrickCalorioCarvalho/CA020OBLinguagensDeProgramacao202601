```python
with open('mensagem.txt','w',encoding='utf-8') as arquivo:
  arquivo.write('Olá mundo!\n')
  arquivo.write('Esta é um exemplo de gravação de arquivos.')

print('Arquivo criado e gravado com sucesso!!')
```

```python
with open('mensagem.txt','r',encoding='utf-8') as arquivo:
  conteudo = arquivo.read()
  print('Conteúdo do arquio:')
  print(conteudo)
```
